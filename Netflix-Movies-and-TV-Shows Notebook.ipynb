The description column in the input data may contain punctuations, non english words, named entities which cannot be used to group the data.
So the above mentioned ambiguities needs to be removed from the description data.
Here SpaCy is used for Natural Language Processing.

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
for token in doc:
    print(token.text, token.pos_, token.dep_)
```
