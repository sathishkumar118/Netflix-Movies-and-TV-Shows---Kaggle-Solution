```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
data = pd.read_csv("../input/netflix-shows/netflix_titles.csv")
data.head()
```

The description column in the input data may contain punctuations, non english words, named entities which cannot be used to group the data.
So the above mentioned ambiguities needs to be removed from the description data.
Here SpaCy is used for Natural Language Processing.

```python
import re
import string
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from collections import Counter
descs = pd.DataFrame(data.description)
descs_arr = np.array(descs)
table = str.maketrans('', '', string.punctuation)
nlp = spacy.load("en_core_web_sm")
#descs_arr
i=0
final_concat = []
for desc in descs_arr:
    words=str(desc).split()
    stripped = [w.translate(table).lower() for w in words]
    stripped = [w for w in stripped if not w in STOP_WORDS]
    words = ' '.join(stripped)
    lem = nlp(words)
    stripped = [w.lemma_ for w in lem]
    final_concat = final_concat + stripped
word_freq = Counter(final_concat).most_common(100)
```
